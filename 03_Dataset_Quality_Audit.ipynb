```python
import json

with open(
    "master_dataset.json",
    "r",
    encoding="utf-8"
) as f:

    dataset = json.load(f)

print("="*60)
print("DATASET LOADED SUCCESSFULLY")
print("="*60)

print()

print("Total Records :", len(dataset))
```

```python
# ==========================================
# DATASET STATISTICS
# ==========================================

lengths = []

for record in dataset:
    lengths.append(len(record["transcript"].split()))

print("="*60)
print("DATASET STATISTICS")
print("="*60)

print()

print("Total Records   :", len(dataset))
print("Average Length  :", round(sum(lengths)/len(lengths),2))
print("Minimum Length  :", min(lengths))
print("Maximum Length  :", max(lengths))
```

```python
from collections import Counter

speaker_counter = Counter()

for record in dataset:
    speaker_counter[record["speaker"]] += 1

print("="*60)
print("SPEAKER DISTRIBUTION")
print("="*60)

for speaker, count in speaker_counter.items():
    print(f"{speaker} : {count}")
```

```python
transcripts = [
    record["transcript"].strip().lower()
    for record in dataset
]

unique = len(set(transcripts))
duplicates = len(transcripts) - unique

print("="*60)
print("DUPLICATE ANALYSIS")
print("="*60)

print()

print("Total Records :", len(transcripts))
print("Unique Records:", unique)
print("Duplicates    :", duplicates)
```
